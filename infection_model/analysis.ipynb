```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PdGrid_model_step_batch_run_data.csv')
```

```python
len(df)/51
```

```python
df.head()
```

```python
steps = list(range(1,52))*60
```

```python
df['steps'] = steps
```

```python
df.head()
```

```python
def plot_result(key, name):
    plt.figure(dpi = 100)
    plt.plot(list(range(1,52)), 
             df[(df.payoffs == key) & (df.schedule_type == 'Sequential')].groupby('steps').mean().cooperating_agents, 
             label = 'Sequential')
    plt.plot(list(range(1,52)), 
             df[(df.payoffs == key) & (df.schedule_type == 'Random')].groupby('steps').mean().cooperating_agents,
             label = 'Random')
    plt.plot(list(range(1,52)), 
             df[(df.payoffs == key) & (df.schedule_type == 'Simultaneous')].groupby('steps').mean().cooperating_agents,
             label = 'Simultaneous')
    plt.xlabel('steps')
    plt.ylabel('cooperating agents')
    plt.title(f'(D,C):{key}')
    plt.legend()
    plt.savefig(f'{name}.jpg')
```

```python
plot_result(0.5,'05')
```

```python
plot_result(0.9, '09')
```

```python
plot_result(1, '1')
```

```python
plot_result(1.6, '16')
```
